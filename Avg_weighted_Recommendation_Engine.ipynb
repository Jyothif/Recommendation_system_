```python
movies_cleaned_df = movies_df_merge.drop(columns = ["homepage","title_x","title_y","status","production_countries"])
```

### Using Weighted average for each movie's Average rating

```python
# calculating all the components based on the above formula
```

```python
v =  movies_cleaned_df['vote_count']
R =  movies_cleaned_df['vote_average']
c =  movies_cleaned_df['vote_average'].mean()
m =  movies_cleaned_df['vote_count'].quantile(0.70)
```

```python
movies_cleaned_df['weighted_avg'] = ((R*v)+(c*m))/(v+m)
```

```python
movies_cleaned_df.head()
```

```python
movie_sorted_ranking = movies_cleaned_df.sort_values("weighted_avg",ascending=False)
```

```python
movie_sorted_ranking[['original_title','vote_count','vote_average','weighted_avg','popularity']].head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
weight_avg = movie_sorted_ranking.sort_values('weighted_avg',ascending =False)

#plotting
plt.figure(figsize=(10,5))
axis1 = sns.barplot(x=weight_avg['weighted_avg'].head(20), y = weight_avg['original_title'].head(20),data = weight_avg)
plt.xlim(4,10)
plt.title("Best Movies by average votes", weight = 'bold')
plt.xlabel('weighted Avg score', weight = 'bold')
plt.ylabel('Movie Title', weight = 'bold')
plt.savefig('best_movies.png')
```

```python
popularity = movie_sorted_ranking.sort_values('popularity',ascending = False)
plt.figure(figsize=(12,6))
ax=sns.barplot(x=popularity['popularity'].head(20), y=popularity['original_title'].head(20), data=popularity)

plt.title('Most Popular by votes', weight = 'bold')
plt.xlabel('Score of Popularity', weight = 'bold')
plt.ylabel('Movie title', weight = 'bold')
plt.savefig('best_popular_movies.png')      
```

### Recommendation based on scaled weighted average and popularity score(Priority is given 50% to weighted_avg and Popularity score)


```python
from sklearn.preprocessing import MinMaxScaler

scaling = MinMaxScaler()
movie_scaled_df = scaling.fit_transform(movies_cleaned_df[['weighted_avg',"popularity"]])
movie_normalized_df = pd.DataFrame(movie_scaled_df,columns = ['weighted_avg','popularity'])
```

```python
movie_normalized_df.head()
```

```python
movies_cleaned_df[['normalized_weight_average','normalize_popularity']] = movie_normalized_df
```

```python
movies_cleaned_df.head()
```

```python
movies_cleaned_df['score'] = movies_cleaned_df['normalized_weight_average']*0.5 + movies_cleaned_df['normalize_popularity']*0.5
movies_scored_df = movies_cleaned_df.sort_values(['score'],ascending= False)
movies_scored_df[['original_title','normalize_popularity',"score"]].head(20)
```

```python
scored_df = movies_cleaned_df.sort_values('score', ascending=False)

plt.figure(figsize=(16,6))

ax = sns.barplot(x=scored_df['score'].head(20), y=scored_df['original_title'].head(20), data=scored_df, palette='deep')

#plt.xlim(3.55, 5.25)
plt.title('Best Rated & Most Popular Blend', weight='bold')
plt.xlabel('Score', weight='bold')
plt.ylabel('Movie Title', weight='bold')

plt.savefig('scored_movies.png')
```
